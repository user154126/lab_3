# Задание №1. Построение простого графика.

Напишите код, который строит график функции y = x2 на интервале [−10, 10]. Добавьте заголовок, подписи осей и сетку.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-10, 10, 50)
y = x**2 

plt.plot(x, y, label=r"$y = x^2$", color="b")

plt.title("График функции $y = {x^2}$")
plt.xlabel("x")
plt.ylabel("y")

plt.grid(True)

plt.show()
```

# Задание №2. Настройка стилей и цветов.

Постройте три линии на одном графике:

1. y = x (синяя, пунктирная линия),
2. y = $x^2$ (зеленая, штрих-пунктирная линия),
3. y = $x^3$ (красная, сплошная линия).

Добавьте легенду и сделайте оси одинакового масштаба.

```python
x = np.linspace(-10, 10, 50)
y1 = x
y2 = [i**2 for i in x]
y3 = [i**3 for i in x]

plt.figure(figsize=(5, 5))
plt.plot(x, y1, 'b--', label="y = x")
plt.plot(x, y2, 'g-.', label=r"$y = {x^2}$")
plt.plot(x, y3, 'r-', label=r"$y = {x^3}$")

plt.title(r"Зависимости: $y1 = x, y2 = {x^2}, y3 = {x^3}$")
plt.xlabel("x")
plt.ylabel("y1, y2, y3")
plt.legend()
plt.grid()

plt.show()
```

# Задание №3. Использование различных типов графиков.

# Задание №4. Гистограмма распределения.

Сгенерируйте 1000 случайных чисел из нормального распределения с
параметрами μ = 0, σ = 1 и постройте их гистограмму с 30 бинами. Добавьте
вертикальную линию в среднем значении.

```python
np.random.seed(122) 
data = np.random.normal(0, 1, 1000)

counts, bins = np.histogram(data, bins=30)

plt.figure(figsize=(6, 5))
plt.bar(bins[:-1], counts, width=np.diff(bins), color='red', edgecolor='black', alpha=0.7)

plt.axvline(np.mean(data), color='black', linestyle='dashed', linewidth=2, label='Среднее значение')

plt.title("Гистограмма нормального распределения")
plt.xlabel("Значения")
plt.ylabel("Частоты")
plt.legend()
plt.grid()

plt.show()
```

# Задание №7. Трехмерный график поверхности.

Используя mpl_toolkits.mplot3d , постройте 3D-график функции $z = \sin\left(\sqrt{x^2 + y^2}\right)$
на сетке значений x, y в диапазоне [−5, 5].

```python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.sin(np.sqrt(X**2 + Y**2))

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, Z, cmap='inferno')

plt.show()
```

### Задание №3. Площадь под экспоненциальной функцией.

Найдите определенный интеграл:

$f(x) = e^{2x}$

на отрезке [0, 1].

1. Построить график подинтегральной функции.
2. Вычислить площадь под кривой на заданном отрезке как значение
определенного интеграла.
3. Закрасить область под графиком, чтобы визуализировать интеграл.

```python
def f(x):
    return np.exp(2 * x)

x = np.linspace(0, 1, 100)
y = f(x)

integral_value = (np.exp(2*1) - np.exp(2*0)) / 2

plt.plot(x, y, label=r"$f(x) = e^{2x}$")

plt.fill_between(x, y, color="grey", alpha=0.5, label=f"Площадь ≈ {integral_value:.4f}")

plt.xlabel("x")
plt.ylabel("f(x)")
plt.title(r"Площадь под кривой $f(x) = e^{2x}$ на [0,1]")
plt.legend()
plt.grid(alpha=0.5)

plt.show()
```

## Задача на построение 3D-графика с помощью Matplotlib

### Задание №4. Полусфера.

Постройте график полусферы: $f(x, y) = \sqrt{1 - x^2 - y^2}$ в области $x^2 + y^2 <= 1$.


1. Построить трехмерный график функции f(x, y) в заданных пределах.
2. Использовать библиотеку Matplotlib для визуализации.
3. Оформить график: добавить заголовок, подписи осей и цветовую карту (если
уместно).

Дополнительные требования:
1. Используйте plot_surface() для плотных поверхностей.
2. Используйте plot_wireframe() для каркасных моделей.
3. Настроите цветовую карту через cmap .
4. Добавьте подписи осей и заголовки.

```python
x = np.linspace(-1, 1, 50)
y = np.linspace(-1, 1, 50)
X, Y = np.meshgrid(x, y)
Z = np.sqrt(np.maximum(0, 1 - X**2 - Y**2))

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(X, Y, Z, cmap="inferno", edgecolor='black', alpha=0.95)

ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel(r"$\sqrt{1 - x^2 - y^2}$")
ax.set_title(r"Полусфера: $f(x, y) = \sqrt{1 - x^2 - y^2}$")

plt.show()
```
